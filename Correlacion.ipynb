```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def Correlacion(archivo):
    Cant_J = len(archivo[:,0])
    Js = []
    Taus = []
    Correl = []
    for j in range (Cant_J):
        Js.append(archivo[j,0]) # creo el vector de 1/J's
        Taus.append(archivo[j,1]) #creo el vector con los taus     
    
    Max = max(Taus)
    print('El valor máximo de tau es %f' %Max)
    plt.figure(figsize = (9,7))
    plt.plot(Js,Taus,'mo-',fillstyle = 'none')
    plt.xlabel('J*', fontsize = 15);
    plt.ylabel(r'$\tau $', fontsize = 15);
    plt.grid();
    plt.show();
    
    plt.figure(figsize = (9,7))
    for i in range (0,Cant_J,15):
        plt.plot(archivo[i,2:],'o-',fillstyle = 'none', label = 'J = %f' %archivo[i,0])
    plt.xlabel('k', fontsize = 15);
    plt.ylabel('C(k)', fontsize = 15);
    plt.axis([0,20,-0.65,1]);
    plt.legend();
    plt.grid();
    plt.show();
```
